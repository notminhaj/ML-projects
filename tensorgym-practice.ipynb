```python
import torch

A = torch.tensor([[1, 0, 2],
                  [0, 0, 0],
                  [1, 2, 3]])

def drop_column(A: torch.tensor) -> torch.tensor:
    x, y = A.shape
    zero_counts = []

    # Count the number of zeros in each column
    for i in range(y):
        zero_counts.append(torch.sum(A[:, i] == 0))
        print(zero_counts)

drop_column(A)
```

```python
import torch

# Example 2D tensor
A = torch.tensor([[0, 0, 0], 
                  [1, 2, -1]])

# Get the maximum values of each row
max_values, max_indices = torch.max(A, dim=1)

max_values = max_values.unsqueeze(0)
max_values = max_values.t()

maxSubtracted = A - max_values
print(maxSubtracted)

exp = torch.exp(maxSubtracted)
print(exp)

expSum = torch.sum(exp, dim=1)
print(expSum)

expSum = expSum.unsqueeze(0)
expSum = expSum.t()

results = exp/expSum

results
```

```python
#Batch norm!!!!!!
data = torch.tensor([[1.0, 4],
                    [2, 5],
                    [3, 6]])
epsilon = 0.00001

dataLen = data.shape[1]
mean = torch.mean(data, dim=0, keepdim=True)
variance = torch.sum((data-mean)**2, dim=0)/dataLen
#var = torch.var(data, dim=0, unbiased=True)
print(dataLen)
print(mean)
print(variance)

result = (data-mean)/((variance + epsilon)**0.5)
print(result)
```
